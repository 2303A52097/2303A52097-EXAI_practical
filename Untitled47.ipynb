```python
# Step 2: Data Preprocessing

import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load dataset
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

# ----- 1Ô∏è‚É£ Drop unnecessary column -----
df.drop(columns=["Person ID"], inplace=True)

# ----- 2Ô∏è‚É£ Handle 'Blood Pressure' column -----
# Split "Blood Pressure" into two numeric columns: Systolic and Diastolic
bp = df["Blood Pressure"].str.split("/", expand=True)
df["Systolic_BP"] = bp[0].astype(int)
df["Diastolic_BP"] = bp[1].astype(int)
df.drop(columns=["Blood Pressure"], inplace=True)

# ----- 3Ô∏è‚É£ Encode categorical columns -----
label_cols = ["Gender", "Occupation", "BMI Category", "Sleep Disorder"]
le = LabelEncoder()

for col in label_cols:
    df[col] = le.fit_transform(df[col])

# ----- 4Ô∏è‚É£ Separate features (X) and target (y) -----
X = df.drop(columns=["Sleep Disorder"])
y = df["Sleep Disorder"]

# ----- 5Ô∏è‚É£ Scale numerical features -----
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for readability
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# ----- 6Ô∏è‚É£ Display results -----
print("‚úÖ Data Preprocessing Completed Successfully!\n")
print("üîπ Features (X) after preprocessing:")
print(X_scaled.head(), "\n")

print("üîπ Target (y):")
print(y.head())
```

```python
# Step 3: Model Training and Evaluation

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings("ignore")

# ------------------------------
# Load and preprocess dataset
# ------------------------------
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

# Drop ID column
df.drop(columns=["Person ID"], inplace=True)

# Split Blood Pressure into two columns
bp = df["Blood Pressure"].str.split("/", expand=True)
df["Systolic_BP"] = bp[0].astype(int)
df["Diastolic_BP"] = bp[1].astype(int)
df.drop(columns=["Blood Pressure"], inplace=True)

# Encode categorical columns
label_cols = ["Gender", "Occupation", "BMI Category", "Sleep Disorder"]
le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])

# Split features and target
X = df.drop(columns=["Sleep Disorder"])
y = df["Sleep Disorder"]

# Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# ------------------------------
# Define models
# ------------------------------
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=300, max_depth=10, random_state=42),
    "Support Vector Machine": SVC(kernel='rbf', random_state=42),
    "Neural Network (MLP)": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

# ------------------------------
# Train & Evaluate all models
# ------------------------------
results = []

print("üîπ Training and Evaluating Models...\n")

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results.append((name, acc))
    print(f"‚úÖ {name} Accuracy: {acc*100:.2f}%")
    print(classification_report(y_test, y_pred))
    print("-" * 60)

# ------------------------------
# Compare Model Performances
# ------------------------------
results_df = pd.DataFrame(results, columns=["Model", "Accuracy"]).sort_values(by="Accuracy", ascending=False)
print("\nüèÜ Model Accuracy Comparison:")
print(results_df)
```

```python
# Step 4: Hyperparameter Tuning (Grid Search)

import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
import warnings
warnings.filterwarnings("ignore")

# ------------------------------
# Load and preprocess dataset
# ------------------------------
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
df.drop(columns=["Person ID"], inplace=True)

# Split Blood Pressure into two columns
bp = df["Blood Pressure"].str.split("/", expand=True)
df["Systolic_BP"] = bp[0].astype(int)
df["Diastolic_BP"] = bp[1].astype(int)
df.drop(columns=["Blood Pressure"], inplace=True)

# Encode categorical columns
label_cols = ["Gender", "Occupation", "BMI Category", "Sleep Disorder"]
le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])

# Split features and target
X = df.drop(columns=["Sleep Disorder"])
y = df["Sleep Disorder"]

# Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# ------------------------------
# 1Ô∏è‚É£ Random Forest Tuning
# ------------------------------
rf = RandomForestClassifier(random_state=42)

rf_params = {
    "n_estimators": [100, 200, 300, 500],
    "max_depth": [5, 10, 15, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

rf_grid = GridSearchCV(estimator=rf, param_grid=rf_params, cv=5, n_jobs=-1, scoring='accuracy')
rf_grid.fit(X_train, y_train)

print("üèÜ Best Random Forest Parameters:")
print(rf_grid.best_params_)
print(f"Best Cross-Validation Accuracy: {rf_grid.best_score_ * 100:.2f}%\n")

# Evaluate on test data
rf_best = rf_grid.best_estimator_
y_pred_rf = rf_best.predict(X_test)
print("‚úÖ Random Forest Test Accuracy:", accuracy_score(y_test, y_pred_rf) * 100)
print(classification_report(y_test, y_pred_rf))


# ------------------------------
# 2Ô∏è‚É£ XGBoost Tuning
# ------------------------------
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

xgb_params = {
    "n_estimators": [100, 200, 300],
    "learning_rate": [0.01, 0.05, 0.1, 0.2],
    "max_depth": [3, 5, 7, 9],
    "subsample": [0.7, 0.8, 1.0],
    "colsample_bytree": [0.7, 0.8, 1.0]
}

xgb_grid = GridSearchCV(estimator=xgb, param_grid=xgb_params, cv=5, n_jobs=-1, scoring='accuracy')
xgb_grid.fit(X_train, y_train)

print("\nüèÜ Best XGBoost Parameters:")
print(xgb_grid.best_params_)
print(f"Best Cross-Validation Accuracy: {xgb_grid.best_score_ * 100:.2f}%\n")

# Evaluate on test data
xgb_best = xgb_grid.best_estimator_
y_pred_xgb = xgb_best.predict(X_test)
print("‚úÖ XGBoost Test Accuracy:", accuracy_score(y_test, y_pred_xgb) * 100)
print(classification_report(y_test, y_pred_xgb))
```

```python
# Step 4: Hyperparameter Tuning + Graphs

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import warnings
warnings.filterwarnings("ignore")

# ------------------------------
# Data Preprocessing (same as before)
# ------------------------------
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
df.drop(columns=["Person ID"], inplace=True)

bp = df["Blood Pressure"].str.split("/", expand=True)
df["Systolic_BP"] = bp[0].astype(int)
df["Diastolic_BP"] = bp[1].astype(int)
df.drop(columns=["Blood Pressure"], inplace=True)

label_cols = ["Gender", "Occupation", "BMI Category", "Sleep Disorder"]
le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])

X = df.drop(columns=["Sleep Disorder"])
y = df["Sleep Disorder"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# ------------------------------
# 1Ô∏è‚É£ Random Forest (quick grid)
# ------------------------------
rf = RandomForestClassifier(random_state=42)
rf_params = {
    "n_estimators": [100, 200],
    "max_depth": [5, 10, None],
}
rf_grid = GridSearchCV(rf, rf_params, cv=3, n_jobs=-1, scoring="accuracy")
rf_grid.fit(X_train, y_train)
rf_best = rf_grid.best_estimator_
rf_acc = accuracy_score(y_test, rf_best.predict(X_test))

# ------------------------------
# 2Ô∏è‚É£ XGBoost (quick grid)
# ------------------------------
xgb = XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42)
xgb_params = {
    "n_estimators": [100, 200],
    "max_depth": [3, 5],
    "learning_rate": [0.05, 0.1]
}
xgb_grid = GridSearchCV(xgb, xgb_params, cv=3, n_jobs=-1, scoring="accuracy")
xgb_grid.fit(X_train, y_train)
xgb_best = xgb_grid.best_estimator_
xgb_acc = accuracy_score(y_test, xgb_best.predict(X_test))

# ------------------------------
# üìà Compare Accuracy in Graph
# ------------------------------
models = ["Random Forest", "XGBoost"]
accuracies = [rf_acc * 100, xgb_acc * 100]

plt.figure(figsize=(6,4))
sns.barplot(x=models, y=accuracies, palette="viridis")
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy (%)")
plt.ylim(0, 100)
for i, acc in enumerate(accuracies):
    plt.text(i, acc + 1, f"{acc:.2f}%", ha="center", fontsize=11)
plt.show()

# ------------------------------
# üîç Confusion Matrices
# ------------------------------
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
cm_rf = confusion_matrix(y_test, rf_best.predict(X_test))
cm_xgb = confusion_matrix(y_test, xgb_best.predict(X_test))

sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title("Random Forest Confusion Matrix")

sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Greens', ax=axes[1])
axes[1].set_title("XGBoost Confusion Matrix")

plt.tight_layout()
plt.show()

# ------------------------------
# üåü Feature Importance Graphs
# ------------------------------
rf_imp = pd.Series(rf_best.feature_importances_, index=X.columns).sort_values(ascending=False)
xgb_imp = pd.Series(xgb_best.feature_importances_, index=X.columns).sort_values(ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

rf_imp.head(10).plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title("Top 10 Important Features (Random Forest)")
axes[0].set_ylabel("Importance Score")

xgb_imp.head(10).plot(kind='bar', ax=axes[1], color='limegreen')
axes[1].set_title("Top 10 Important Features (XGBoost)")
axes[1].set_ylabel("Importance Score")

plt.tight_layout()
plt.show()

# ------------------------------
# üèÅ Print Summary
# ------------------------------
print("\nüèÜ Model Accuracy Summary:")
print(f"Random Forest: {rf_acc*100:.2f}% (Best Params: {rf_grid.best_params_})")
print(f"XGBoost: {xgb_acc*100:.2f}% (Best Params: {xgb_grid.best_params_})")
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# ------------------------------
# üîπ ROC Curve (Multi-class)
# ------------------------------

# Binarize the target for multi-class ROC
y_test_bin = label_binarize(y_test, classes=sorted(y.unique()))
n_classes = y_test_bin.shape[1]

# Get prediction probabilities
y_score_rf = rf_best.predict_proba(X_test)
y_score_xgb = xgb_best.predict_proba(X_test)

# Plot ROC for each class
plt.figure(figsize=(10,6))

for i in range(n_classes):
    fpr_rf, tpr_rf, _ = roc_curve(y_test_bin[:, i], y_score_rf[:, i])
    fpr_xgb, tpr_xgb, _ = roc_curve(y_test_bin[:, i], y_score_xgb[:, i])

    plt.plot(fpr_rf, tpr_rf, linestyle='--', label=f'RF Class {i} (AUC={auc(fpr_rf, tpr_rf):.2f})')
    plt.plot(fpr_xgb, tpr_xgb, label=f'XGB Class {i} (AUC={auc(fpr_xgb, tpr_xgb):.2f})')

plt.plot([0,1], [0,1], color='grey', linestyle=':')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Multi-class')
plt.legend()
plt.show()
```
